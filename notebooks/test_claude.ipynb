````python
test_prompt = "You will be presented with a context-free grammar in Chomsky normal form and a string which may or may not be in the language defined by the given grammar. Your job is to determine whether or not the grammar generates the provided string. You can use any reasoning strategy you like, but you must end your response with either 'Yes' (if the string is generated by the grammar) or 'No' (if it isn't.)\n\nGrammar: ```S -> NT2 NT2\nS -> NT0 NT1\nS -> NT3 NT5\nNT0 -> NT2 NT5\nNT3 -> NT2 NT0\nNT2 -> NT0 NT2\nNT0 -> NT2 NT0\nNT0 -> 't4'\nNT1 -> 't0'\nNT5 -> 't4'\nNT3 -> 't5'\nNT5 -> 't5'\nNT2 -> 't2'\nNT0 -> 't2'\nNT3 -> 't4'\n\nHere is the string you need to evaluate:\n\nString: `t4 t2 t5 t2 t4 t2 t4 t2 t4 t2 t4 t4 t4 t2 t2 t5 t2 t5 t4 t4 t2`.\n\nRemember, end your response with either 'Yes' or 'No'."  # noqa: E501

test_prompt_2 = "You will be presented with a context-free grammar in Chomsky normal form and a string which may or may not be in the language defined by the given grammar. Your job is to determine whether or not the grammar generates the provided string. You can use any reasoning strategy you like, but you must end your response with either 'Yes' (if the string is generated by the grammar) or 'No' (if it isn't.)\n\nGrammar: ```S -> NT2 NT2\nS -> NT0 NT1\nS -> NT3 NT5\nNT0 -> NT2 NT5\nNT3 -> NT2 NT0\nNT2 -> NT0 NT2\nNT0 -> NT2 NT0\nNT0 -> 't4'\nNT1 -> 't0'\nNT5 -> 't4'\nNT3 -> 't5'\nNT5 -> 't5'\nNT2 -> 't2'\nNT0 -> 't2'\nNT3 -> 't4'\n\nHere are some examples of some positive and negative samples from the provided grammar:\n- `t2 t5 t2 t2 t4 t5` -> Yes\n- `t5 t5 t2 t4 t5 t2 t5 t4 t5 t4 t2 t5 t5 t2 t5 t4 t4 t5 t2 t5 t4 t4 t2 t2 t5 t5 t4 t2 t4 t4 t0 t2 t4 t4 t4 t2 t4 t4` -> No\n\nHere is the string you need to evaluate:\n\nString: `t2 t2 t2 t4 t2 t5 t2 t2 t2 t4 t2 t2 t5 t2`.\n\nRemember, end your response with either 'Yes' or 'No'."  # noqa: E501
````

````python
client = anthropic.Client()

message = client.messages.create(
    model="claude-3-haiku-20240307",
    max_tokens=1024,
    temperature=0,
    messages=[
        {"role": "user", "content": test_prompt_2},
    ],
    metadata={"sample": "t2 t2 t2 t4 t2 t5 t2 t2 t2 t4 t2 t2 t5 t2"},
)
````
